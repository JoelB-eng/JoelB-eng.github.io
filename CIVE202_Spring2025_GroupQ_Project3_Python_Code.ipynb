# Define all libraries that will be used.
This section defines the libraries that will be used to generate the box plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Define the Dataset
This portion of the code defines the dataset that will be used to isolate variables for the graph

```python
nhts_data = pd.read_csv('NHTS.csv')

nhts_data.columns
```

# Create the Box Plot
This section creates the box plot using the income brackets and number of cars per household. It customizes the graph to fit the variables that are being used in this graph.

```python
# Creating the Box plot

plt.figure(figsize=(12,6))

boxplot = nhts_data.boxplot(column='vehicles_per_household', by='household_income')

# Customizing the Box Plot
plt.xticks(rotation = 90)

plt.xlabel('Househole Income')
plt.ylabel('Vehicles Per Household')

plt.title('Boxplot of Household Vehicles by Income')

# Displaying the Box Plot
plt.suptitle('')
plt.show()
```

# Creating a Histogram
This section creates a histogram based off of the distribution of vehicle age

```python
##Tell Python you are plotting a figure and how big you want the figure to be
plt.figure(figsize= (8,8))

## Tell Python what type of plot you want.  Here, .hist() creates a histogram
#plt.hist(nhts_data.vehicle_age)
sns.histplot(nhts_data['vehicle_age'], bins=15, kde=True, color='red')

# Then let's label.  You'll notice you can change the fontsize of all your labels!
plt.title('Vehicle Age Distribution', fontsize = 16)
plt.xlabel('Age in Years', fontsize = 14)
plt.ylabel('Count', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
```
